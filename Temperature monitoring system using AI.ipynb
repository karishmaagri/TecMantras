```python
import time
import random
```

```python
#Simulate real-time temperature data
locations = ['Mumbai', 'Ahmedabad', 'Pune', 'Vapi', 'Bangalore']

print("Streaming temperature data.....")

while True:
    data = {
        'location': random.choice(locations),
        'temperature': round(random.uniform(20, 40), 2)
    }
    print(f"Data: {data}")
    with open('temperature_data.txt', 'a') as file:
        file.write(f"{data['location']},{data['temperature']}\n")
    time.sleep(5) #Generate data every 2 seconds
```
